```python
# tạo một biểu đồ phân phối cho các loài chim cánh cụt
import seaborn as sns
import pandas as pd

penguins= pd.read_csv("penguins.csv")

sns.displot(data=penguins, x="flipper_length_mm", hue="species", multiple="layer", alpha = 1)

# khi bỏ multiple="stack" thì mặc định là multiple="layer" và alpha lúc này sẽ được tự động điều chỉnh 
# để có thể nhìn thấy các lớp chồng lên nhau
# nếu không muốn điều chỉnh alpha thì có thể sử dụng multiple="stack" để chồng lên nhau
# khi sử dụng multiple="stack" thì alpha sẽ được điều chỉnh để có thể nhìn thấy các lớp chồng lên nhau
# có thể sử dụng alpha để điều chỉnh độ trong suốt của các lớp chồng lên nhau
# alpha = 1 sẽ cho phép nhìn thấy các lớp chồng lên nhau
# alpha = 0.5 sẽ làm cho các lớp chồng lên nhau trở nên trong suốt hơn
```

```python
sns.displot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack", kind="kde")
# kde là kernel density estimation, là một phương pháp để ước lượng mật độ xác suất của một biến liên tục
# kde là một phương pháp để ước lượng mật độ xác suất của một biến liên tục

# khi không dùng kind thì mặc định là kind="hist" và alpha lúc này sẽ được tự động điều chỉnh
# để có thể nhìn thấy các lớp chồng lên nhau
```

```python
sns.displot(data=penguins, x="flipper_length_mm", hue="species", col="species")

# col là một tham số để chia nhỏ biểu đồ thành nhiều biểu đồ con theo từng giá trị của biến phân loại
# trong trường hợp này là species, mỗi biểu đồ con sẽ có một tiêu đề riêng với mỗi giá trị của biến phân loại
# hue là một tham số để chia nhỏ biểu đồ thành nhiều biểu đồ con theo từng giá trị của biến phân loại
# trong trường hợp này là species, mỗi biểu đồ con sẽ có một màu khác nhau cho mỗi giá trị của biến phân loại
# x là một tham số để xác định biến x của biểu đồ
# flipper_length_mm là một biến x của biểu đồ
# hue là một tham số để chia nhỏ biểu đồ thành nhiều biểu đồ con theo từng giá trị của biến phân loại
# trong trường hợp này là species, mỗi biểu đồ con sẽ có một màu khác nhau cho mỗi giá trị của biến phân loại
# col là một tham số để chia nhỏ biểu đồ thành nhiều biểu đồ con theo từng giá trị của 
```
