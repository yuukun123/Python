```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
g = sns.relplot(data=tips, x="total_bill", y="tip", kind="scatter")
# xy1=(10, 2) là x = 10, y = 2
# Công thức đường thẳng: y - 2 = 0.2 * (x - 10)
# Thay đổi độ dày và kiểu nét của đường thẳng
# Đường thẳng này bắt đầu từ (10, 2) với độ dốc 0.2
# nếu total_bill tăng 1, thì tip tăng 0.2.

# dashes kiểm soát kiểu nét của đường thẳng.
# (5, 2) nghĩa là đường thẳng sẽ được vẽ với các đoạn nét đứt: mỗi đoạn dài 5 đơn vị, sau đó là khoảng trống 2 đơn vị, lặp lại liên tục.
g.ax.axline(xy1=(10, 2), slope=.2, color="b", dashes=(5, 2))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

penguins = sns.load_dataset("penguins")

g = sns.relplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", col="sex", kind="scatter",
    height=4, aspect=1.2 ) # Chiều cao và tỉ lệ khung hình mỗi subplot
# Tùy chỉnh lưới, khoảng cách giữa các plot và margin
g.fig.subplots_adjust(top=0.85, wspace=0.3)  # Căn chỉnh khoảng cách và lề trên

# Thêm tiêu đề lớn cho toàn bộ figure
g.fig.subplots_adjust(top=0.8)  # Hạ thấp suptitle xuống dễ thấy hơn
g.fig.suptitle("Relationship between Flipper Length and Bill Length by Sex")

g.set_axis_labels("Flipper length (mm)", "Bill length (mm)")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Thiết lập phong cách Seaborn với lưới
sns.set_style("whitegrid")  # Phong cách có lưới

penguins = sns.load_dataset("penguins")

# Tạo FacetGrid, chia theo cột 'sex'
g = sns.FacetGrid(data=penguins, col="sex", height=4, aspect=1.3)  # Chiều cao và tỉ lệ khung hình mỗi subplot

# Map scatterplot vào từng grid
g.map_dataframe(sns.scatterplot, x="flipper_length_mm", y="bill_length_mm")

    
# Căn lề và thêm tiêu đề
g.fig.subplots_adjust(top=0.8, wspace=0.3)
g.fig.suptitle("Relationship between Flipper Length and Bill Length by Sex")

# Đặt nhãn trục
g.set_axis_labels("Flipper length (mm)", "Bill length (mm)")

# Hiển thị
plt.show()
```
